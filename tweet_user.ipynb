```python
import pymysql
import json
import pandas as pd
tweets_user = []
userid=[]
with open('./data.txt') as f:
    data = json.loads(f.read())
for line in data:
    try:
        user=line["user"]
        Id=user["id"]
        if Id not in userid:
            userid.append(Id)
            tweets_user.append(user)
    except:
        continue
```
